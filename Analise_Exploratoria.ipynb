```python
# Criando dataframe
df = pd.read_excel('AdventureWorks.xlsx')
```

```python
# Minha receita total
df['Valor Venda'].sum()
```

```python
# Criando coluna de Custo Venda
df['Custo Venda'] = df['Custo Unitário'].mul(df['Quantidade'])
```

```python
df.head()
```

```python
df.shape
```

```python
# Meu custo total
df['Custo Venda'].sum()
```

```python
# Meu lucro
df['Lucro'] = df['Valor Venda'] - df['Custo Venda']
```

```python
df.sample(5)
```

```python
# Meu lucro total
round(df['Lucro'].sum(), 2)
```

### Encontrando o tempo de envio para cada marca

```python
df['Tempo Envio'] = df['Data Envio'] - df['Data Venda']
```

```python
df.sample(1)
```

```python
# Extraindo somente o numero de dias no formato inteiro
df['Tempo Envio'] = df['Tempo Envio'].dt.days
```

```python
df.sample(1)
```

```python
# Verificar a media do Tempo de Envio por Marca
df.groupby('Marca')['Tempo Envio'].mean()
```

```python
# Verifica se há valores faltantes
df.isnull().sum()
```

```python
# Qual o meu lucro por ano e por marca
df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
# Colocando em dataframe o resultado do groupby
lucro_ano = df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum().reset_index()
lucro_ano
```

```python
# Filtrar por data
lucro_ano_2008 = lucro_ano.loc[lucro_ano['Data Venda'] == 2008]
lucro_ano_2008
```

```python
lucro_ano.dtypes
```

```python
# Qual a quantidade total de produtos vendidos ?
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending = False)
```

```python
# Gráfico para produtos vendidos
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending = True).plot.barh(title = 'Total Produtos Vendidos');
plt.xlabel('Total');
plt.ylabel('Produto');
```

```python
# Gráfico para o meu lucro por ano
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum().plot.bar(title='LUCRO x ANO');
plt.xlabel('Ano');
plt.ylabel('Lucro');
plt.xticks(rotation = 'horizontal');
```

```python
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum().plot.pie(title='LUCRO x ANO');
plt.xlabel('Ano');
plt.ylabel('Lucro');
```

```python
# Selecionando dados somente de 2009
df_2009 = df[df['Data Venda'].dt.year == 2009]
```

```python
df_2009.head()
```

```python
# Qual meu lucro por mês no ano de 2009 ? (Gráfico de linha)
df_2009.groupby(df['Data Venda'].dt.month)['Lucro'].sum().plot(title = 'Lucro x Mês');
plt.xlabel('Mes');
plt.ylabel('Lucro');
```

```python
# Gráfico para o lucro total por Marcas em 2009
df_2009.groupby('Marca')['Lucro'].sum().plot.bar(title = 'Lucro x Marca');
plt.xlabel('Marca');
plt.ylabel('Lucro');
plt.xticks(rotation='horizontal');
```
